```
# Standard Deviation computated using Statistics module


import statistics

cost = [10, 12, 14, 15, 16, 16, 17, 18.5, 19, 19.5]
time = [5, 7, 8, 9, 10, 10, 11, 11.5, 12, 13]
units = [19, 21, 22, 23, 24, 24, 25, 26, 27, 28]
costsum = sum(cost)
unitssum = sum(units)

#Standard Error of Cost variables from Mean

costhistderror = []
costlowstderror = []

for i in cost:
  if i >= statistics.mean(cost):
    costhistderror.append(i)
  elif i <= statistics.mean(cost):
      costlowstderror.append(i)
      
#Standard Error of Time variables from Mean

timehistderror = []
timelowstderror = []

for i in time:
  if i >= statistics.mean(time):
    timehistderror.append(i)
  elif i <= statistics.mean(time):
      timelowstderror.append(i)
      
#Standard Error of Units variables from Mean

unitshistderror = []
unitslowstderror = []

for i in units:
  if i >= statistics.mean(units):
    unitshistderror.append(i)
  elif i <= statistics.mean(units):
      unitslowstderror.append(i)

#Cost Output

print ('\033[1m', 'For Cost','\033[0m', '\n')
print ('\tSum=', costsum, '\n', '\tMin=',min(cost), '\n', '\tMax=', max(cost), '\n', '\tMode=', statistics.mode(cost), '\n', '\tMedian=',statistics.median(cost), '\n')
print ('\033[1m' ,'Standard Error of Cost variables from Mean:','\033[0m','\n')
print ('\tMean of cost= $',statistics.mean(cost), '\n')
print ('\tValues higher than Mean:',costhistderror, '\n')
print ('\tValues lower than Mean:',costlowstderror, '\n')
print ('\033[1m' ,'Standard Deviation of Cost:', '\033[0m', statistics.pstdev(cost), '\n')
print ('\033[1m', ('-' * 75),'\033[0m', '\n')

#Time Output

print ('\033[1m','For Time:','\033[0m','\n')
print ('\tSum=',timesum, '\n', '\tMin=',min(time), '\n', '\tMax=', max(time), '\n', '\tMode=', statistics.mode(time),'\n', '\tMedian=', statistics.median(time), '\n')
print ('\033[1m','Standard Error of Time variables from Mean:','\033[0m', '\n')
print ('\tMean of Time=',statistics.mean(time),'mins', '\n')
print ('\tValues higher than Mean:',timehistderror, '\n')
print ('\tValues lower than Mean:',timelowstderror, '\n')
print ('\033[1m''Standard Deviation of Time:', '\033[0m', statistics.pstdev(time), '\n')
print ('\033[1m', ('-' * 75),'\033[0m', '\n')

#Units Output

print ('\033[1m', 'For Units','\033[0m', '\n')
print ('\tSum=',unitssum, '\n', '\tMin=', min(units), '\n', '\tMax=', max(units), '\n', '\tMode=', statistics.mode(units), '\n', '\tmedian=',statistics.median(units), '\n')
print ('\033[1m' ,'Standard Error of Units variables from Mean:','\033[0m','\n')
print ('\tMean of Units=', statistics.mean(units), '\n')
print ('\tValues higher than Mean:',unitshistderror, '\n')
print ('\tValues lower than Mean:',unitslowstderror, '\n')
print ('\033[1m' ,'Standard Deviation of Units:', '\033[0m', statistics.pstdev(units), '\n')
```
